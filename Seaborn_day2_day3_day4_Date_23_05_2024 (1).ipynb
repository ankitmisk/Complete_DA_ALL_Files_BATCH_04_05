```python
# sns.heatmap
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
corr_df  = df.corr(numeric_only=True)
corr_df
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(corr_df,annot=True,fmt='.1f',cmap=sns.color_palette('rainbow'),
            annot_kws={'color':'k','fontsize':8},
           linewidths=10,linecolor='k')
plt.show()
```

```python
described_data = df.describe()
described_data
```

```python
sns.heatmap(described_data)
plt.show()
```

```python
# sns.lmplot
```

```python
sns.lmplot(df,x = 'total_bill',y='tip',hue='sex',col='day')
plt.show()
```
