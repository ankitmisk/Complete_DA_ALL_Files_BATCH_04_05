```python
data1 = {'Id':[101,102,103,104,105],
        'Marks':[23,34,45,67,78]}

data2 = {"Fees":['Yes','Yes','No','No','Yes'],
         'Info Id':[105,103,102,101,104]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)


df1.append(df2,ignore_index=True)
```

```python
data1 = {'Id':[101,102,103,104,105],
        'Marks':[23,34,45,67,78]}

data2 = {"Fees":['Yes','Yes','No','No','Yes'],
         'Info Id':[105,103,102,101,104]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)


df1.append(df2,verify_integrity=True)
```
