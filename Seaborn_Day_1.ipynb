```python
df = sns.load_dataset('flights')
df
```

```python
feb_df = df[df['month'] == 'Feb']
feb_df
```

```python
plt.figure(figsize=(7,3))
sns.lineplot(data = feb_df,x ='year',y = 'passengers') #,hue = 'month',
# style='month',
# size='passengers'.,
# palette = sns.color_palette('magma',1)

plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
pivot_df = df.pivot_table(values='passengers',index='year',columns='month')
pivot_df
```

```python
sns.lineplot(data=pivot_df,markers=True)
```

```python
plt.figure(dpi=1000)
plt.title('Year vs Passengers Count')
sns.lineplot(data=df,x='year',y='passengers',hue='month',markers=True,style='month') #style='month'

plt.annotate("Weak!!", xy=(1955, 220), xytext=(1958, 100),arrowprops=dict(arrowstyle="->"))
plt.annotate("TechVidya!!",xy=(1954,500),xytext=(1954, 500))
plt.savefig('Metro.png')
plt.show()
```
