```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# bar plot

x = ['Python','Java','C','C++','Php','Html']

y = [2345,2345,3456,4567,6543,4567]

plt.bar(x,y,color='g')



plt.title('Language Price')
plt.xlabel('Language')
plt.ylabel('Price')
plt.show()
```

```python
# bar plot

x = ['Python','Java','C','C++','Php','Html']

y = [2345,2345,3456,4567,6543,4567]

plt.figure(figsize=(10,5))

plt.bar(x,y,
        color=['red','blue','green','orange','pink','magenta'],
       alpha = 0.8,edgecolor='k',
       linestyle = '--',
       hatch = 'x',
       label = 'Courses',
       width = 0.8,
#        visible = False,
        linewidth = 2
       )



plt.title('Language Price')
plt.xlabel('Language')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
import seaborn as sns
```

```python
# bar plot

x = ['Python','Java','C','C++','Php','Html']

y = [2345,2345,3456,4567,6543,4567]

plt.figure(figsize=(7,10))

ax = plt.bar(x,y,
        color=sns.color_palette('rainbow'),
       alpha = 0.8,edgecolor='k',
       linestyle = '--',
       hatch = '|',
       label = 'Courses',
       width = 0.8,
#        visible = False,
        linewidth = 2
       )



plt.title('Language Price')
plt.xlabel('Language')
plt.ylabel('Price')
plt.legend()

plt.bar_label(ax,rotation = 90)

plt.tight_layout()
plt.show()
```

```python
# bar plot

x = ['Python','Java','C','C++','Php','Html']

y = [2345,2345,3456,4567,6543,4567]

plt.figure(figsize=(6,3))

ax = plt.bar(x,y,
        color=sns.color_palette('rainbow'),
       alpha = 0.8,edgecolor='k',
       linestyle = '--',
       hatch = '|',
       label = 'Courses',
       width = 0.8,
#        visible = False,
        linewidth = 2
       )



plt.title('Language Price')
plt.xlabel('Language')
plt.ylabel('Price')
plt.legend()

plt.yticks(range(1000,13000,2000))

plt.bar_label(ax,rotation = 90)

plt.tight_layout()
plt.show()
```

```python
x1 = np.linspace(2,12,6)

ax1 = plt.bar(x1,y,width=0.5,color='#CC8FF5',label='2021')
ax2 = plt.bar(x1+0.5,y1,width=0.5,color='#8FF5DD',label = '2022')
ax3 = plt.bar(x1+1,y2,width=0.5,color='#72E733',label = '2023')


plt.bar_label(ax1,rotation = 90,fontsize=6)
plt.bar_label(ax2,rotation = 90,fontsize=6)
plt.bar_label(ax3,rotation = 90,fontsize=6)

plt.yticks(range(1000,13000,2000))
plt.legend()
plt.xticks(x1+0.5,x,rotation=45)
plt.xlabel('Courses')
plt.ylabel('Price')
plt.show()
```
