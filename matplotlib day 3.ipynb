```python
x = ['Noida','Delhi','Faridabad','Haryana','Gurugram']

y = np.random.randint(1000,2000,5)

print(x,y,sep='\n')
```

```python
plt.figure(figsize=(4,4))
explode_data = [0.0]*len(y)

explode_data[-1] = 0.2
plt.pie(y,
        colors = sns.color_palette('rainbow'),
       autopct = '%.2f%%',
       radius = 1.5,
       labels = x,
       explode = explode_data,
       shadow = True,
       pctdistance = 0.5,
        labeldistance=1.1,textprops = {'fontsize':12,"color":'k'})
plt.show()
```

```python
x = ['Noida','Delhi','Faridabad','Haryana','Gurugram']

y1 = np.random.randint(1000,2000,5)

# step: "Literal['pre', 'post', 'mid'] | None" = None
plt.fill_between(x,y1,
                 color='#42D1FC')


index = list(y1).index(max(y1))
plt.yticks(range(500,3000,500))
plt.axhline(max(y1))
plt.axvline(x[index])

plt.text(x=index,y=max(y1),s='Max value')

plt.show()
```

```python
x = ['Noida','Delhi','Faridabad','Haryana','Gurugram']

y1 = np.random.randint(1000,2000,5)

# step: "Literal['pre', 'post', 'mid'] | None" = None
plt.fill_between(x,y1,
                 color='yellow',alpha=0.2)


plt.title('  '.join('City Price Analysis'.upper()))

plt.xlabel('C  I  T  Y')
plt.ylabel('P  R  I  C  E')


index = list(y1).index(max(y1))
plt.yticks(range(500,3000,500))
plt.axhline(max(y1))
plt.axvline(x[index])

plt.text(x=index,y=max(y1),s=f'Max value {max(y1)}')

min_index = list(y1).index(min(y1))
plt.axhline(min(y1))
plt.axvline(x[min_index])

plt.text(x = min_index,y = min(y1), s= f'Min Value {min(y1)}')

plt.show()
```
