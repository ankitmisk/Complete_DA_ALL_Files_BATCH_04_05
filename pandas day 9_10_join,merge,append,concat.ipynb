```python
df1 = pd.DataFrame({'A':[3,5,10,11,12],
           'B':['Ajay','Raman','Lucky','Mohan','Karan']})

df2 = pd.DataFrame({'A':[1,5,11,10,15],
                   'C':['Noida','Delhi','Goa','Ranchi','Faridabad']})
```

```python
pd.merge(df1,df2,on='A',how='right')
```

```python
pd.merge(df1,df2,on='A',how='right',indicator=True)
```

```python
pd.merge(df1,df2,on='A',how='left',indicator=True)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,how='right',indicator=True)
```

```python
pd.merge(df1,df2,on='A',how='right',indicator=True,validate="one_to_one")
pd.merge(df1,df2,on='A',how='right',indicator=True,validate="1:m")
```

```python
import random

d = ['mon','tue','wed','thu','fri','sat','sun']
m = 'Jan	Feb	Mar	Apr	May	Jun	Jul	Aug	Sep	Oct	Nov	Dec'.split('	')
s = ['s','m','l','xl','xxl','xxxl']


Data = {'Days':[random.choice(d).capitalize() for i in range(50)],
       'Month':[random.choice(m) for i in range(50)],
        'Size':[random.choice(s).upper() for i in range(50)],
       'Sales':np.random.randint(500,900,50)}


df = pd.DataFrame(Data)
df.columns
```

```python
df.pivot_table(index='Days',columns='Month',values='Sales',aggfunc='mean')
```

```python
df[df['Month'] == 'Jul']
```

```python
df.pivot_table(index='Days',columns='Month',values='Sales',aggfunc=np.min,fill_value=0)
```
