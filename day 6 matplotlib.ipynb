```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
# distribution of our data:- frequency 
# histtype: "Literal['bar', 'barstacked', 'step', 'stepfilled']"
plt.hist(df['total_bill'],histtype='stepfilled',color='r')
plt.xlabel('Total Bill---------------------->')
plt.ylabel('Count of customers----------------->')
plt.axvline(10,color='r')
plt.axvline(23,color='r')
plt.axhline(8,color='y')
plt.show()
```

```python
y1 = np.random.standard_normal(10000)
y2 = np.random.standard_normal(5000)
y3 = np.random.standard_normal(2500)

plt.hist(y1,color='r',label = 'Total Bill',hatch='o',)
plt.hist(y2,color='g',label = 'Total Bill',hatch='x')
plt.hist(y3,color='b',label = 'Total Bill',hatch='..')

plt.title('Total Tip vs Customer Count')
plt.legend()
plt.show()
```

```python
# plt.stem()
x = np.arange(1,21)
y = np.random.randint(100,200,20)
plt.stackplot(x,y)
```

```python
x = range(10)
y1 = np.random.randint(10,1000,10)
y2 = np.random.randint(10,500,10)
y3 = np.random.randint(10,200,10)

plt.stackplot(x,y1,color='r',labels='Red Bill',baseline='wiggle',alpha=0.4)  #{'zero', 'sym', 'wiggle', 'weighted_wiggle'}
plt.stackplot(x,y2,color='g',labels='Green Bill',baseline='sym',alpha =0.4)
plt.stackplot(x,y3,color='b',labels='Blue Bill',baseline='wiggle',alpha=0.4)

plt.title('Total Tip vs Customer Count')
plt.grid()
plt.legend()
plt.show()
```

```python
x = range(10)
y = np.random.randint(1000,5000,10)
fig,ax = plt.subplots(2,2,sharex=True,sharey=True,dpi=100) #figsize = (10,10)

fig.supxlabel('X label')
fig.supylabel('Y label')

ax[0,0].bar(x,y,color=sns.color_palette('rainbow'))
ax[0,0].set_title('Bar Plot')
ax[0,1].stem(x,y)
ax[0,1].set_title('Stem Plot')
ax[1,0].plot(x,y)
ax[1,0].set_title('Line Plot')
ax[1,1].hist(np.random.standard_normal(10000))
ax[1,1].set_title('Hist Plot')

plt.show()
```
