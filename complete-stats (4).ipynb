**Sampling Technique:**

2. Stratifed sampling:- Where the population (N) is split into non-overlapping groups(strata-layering)
 > EG: Gender survey, where there are no overlap female and male

 > age group survey,profession survey (such as Doc, Eng, etc)

 > Which is not considered are case of programming language known ( in this it can contain  .net,php,python; but person can know both or all language, so this case is not statified samp)
 
3.  Systametic Sampling:- We pick up nth individual 
 > EG: Suppose, we have to survey in mall about the covid, so we considering any nth sample which will be every 8th person in the mall the survey will be conducted for them, this sampling selction is of random selected, it can be any numbers of ours choices. 
 
4. Convenience Sampling:- only those sample of population of interest
 > EG: Survey on DataScience , this survey will be only taken to the person who know about this, we will not survey the docter or any not related to this field regarding this survey because there profession are different .
 
Example 1: For Exit Poll survey which sample are used, sample used is Random sample, because in thta we dont required any specific requirement all ages can be part of survey, just we want to find maximum winner amongst the sample.

Example 2: For RBI to conduct the household  survey, which sampling technique are used, sample used can be stratified sampling were we can see household sample between male and female surveys. we can also can be considering onvenience sampling were only considering women household surveys.

**Variance**

> In this, the variance have, population variance and sample variance

> If variance is high, the data spread is very high  

> Population variance: 

    sigma square = summation of i=1 to N (x of i minus muu(mean)) the whole square and divided by N

> sample variance:

    small s square = summation of i =1 to n(x of i minus xbar(sampleMean)) the whole square and divided by n-1

**Sample Variance**

> n -1, is also called bayesian correction, Degree of freedom, 

> Why sample variance divided by n- 1 ?

> 

**Standard Deviation(S.D)**

> Its square root of variance

> Basically, from the mean how far the element can be.

  EG: S.D are 1.49,2.39,4.71,5.14
  
  and want to find 5, so 5 is in between 4,71 and 5.14,
  
  So, 1.5 S.D to the mean

# Emperical Formula
> 
> Rule of 68-95-97.7% -- -- 68% lies under the 1S.D to left and right
> 
> Considering mean is 4 and standard deviation as 1
> 
> if we take 4.5 fall in terms of standard deviation, it will fall 0.5 S.D to right

> So, To solve this we need Z Score.
> 
> ZScore = xi - mean/standard deviation

**> To solve for 4.75, ZScore = 4.75-4/1 = 0.75 S.D to the Right, as this is positive value, so it is right.**

> And calculate ZScore of each, we get standard deviations value of each.

# **Normalization**

> To make data set value's from between 0 and 1
> Convert all values into 0 to 1
> 
> To Do Normalization, there is concept of MinMax Scaler
> 
> By defining lower and upper bond we can convert the data between them
> 
> Uses, Image/object Detection, in this cases every images have the pixels, suppose w ehave 4x4 images, each and every pixels box range from 0 - 255, Before start training, we can use minmax scaler and its converted into 0 to 1, we can also divide each pixel by 255 that will convert it to 0 to 1

```python
df = pd.DataFrame([
                   [180000, 110, 18.9, 1400], 
                   [360000, 905, 23.4, 1800], 
                   [230000, 230, 14.0, 1300], 
                   [60000, 450, 13.5, 1500]], 
    
                   columns=['Col A', 'Col B',
                            'Col C', 'Col D'])
  
# view data
display(df)
```

```python
df
```

```python
X_min = df['Col A'].min()
X_max = df['Col A'].max()

df['Col A Normal']  = (df['Col A'] - X_min)/(X_max-X_min)
```

```python
df
```

```python
df.describe()
```

```python
df
```

```python
min_c = df['Col C'].min()
max_c = df['Col C'].max()

df['Col C'] = (df['Col C'] - min_c)/(max_c - min_c)
```

```python
# Sklearn: Ordinal_Encoder
import random
import pandas as pd
import numpy as np
```

```python
all_sizes = ['x','xl','xxl','xxxl','l','m','s']

data = {'Shirt Size':[random.choice(all_sizes) for i in range(20)],
       'Price':np.random.randint(100,500,20)}

df = pd.DataFrame(data)
df
```

```python
from sklearn.preprocessing import OrdinalEncoder

size_ranking = [['s','m','l','x','xl','xxl','xxxl']]
```

```python
aadil = OrdinalEncoder(categories=size_ranking)
```

```python
df['shirt_ranks'] = aadil.fit_transform(df[['Shirt Size']]).astype(int)
```

```python
df
```

```python
map_rank = dict(zip(size_ranking[0],range(len(size_ranking[0]))))
map_rank
```

```python
df['Shirst_Size_Map'] = df['Shirt Size'].map(map_rank)
```
