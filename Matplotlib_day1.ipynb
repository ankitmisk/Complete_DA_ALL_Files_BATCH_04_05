```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
from datetime import datetime
# data1 = [1,23,45,5,32,4,34,6,56,4,3,24]

# data2 = [100,34,45,67,32,34,56,77,35,12,23,34,56,67]
c = 1
while True:

    data1 = np.random.randint(100,500,12)
    
    cur_date = datetime.now()
    
    plt.figure(figsize=(7,3))
    plt.plot(data1,
             color = 'b',
            marker = 'o',
            linestyle = '--',
            linewidth = 1,
            label = 'Blue Line')

    plt.grid(alpha = 0.2,color='r')

    plt.title(f'Blue Line Metro            Image {c}\n{cur_date}'.upper(),color='b',fontsize = 15)
    

    plt.xlabel('Months',color='g',fontsize = 7)

    plt.ylabel('Passengers',color='r',fontsize = 7)

    months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

    plt.xticks(range(len(months)),months,
               rotation = 45,
               color = 'g',
              fontsize = 7)


    plt.legend()
    
    plt.tight_layout()
    plt.savefig(f"C:\\Users\\Lenovo\\Desktop\\bluelinemetrodata\\image_{c}.jpg",dpi = 1000)
    
    plt.show()
    c+=1
    display(clear=True)
```
